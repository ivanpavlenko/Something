## Physical model
The magnetohydrodynamic instability growth on the concave wall is investigated. The physical model needs to be described to rationalize the validity of using *single-fluid equations* to solve the problem of instability. The single-fluid model was thoroughly considered by *Chandrasekhar* in his book *"Hydrodynamic and hydromagnetic stability"*. Also, the principle of the single-fluid model constructing was described by *Freidberg* in his paper *"Ideal magnetohydrodynamic theory of magnetic fusion systems", July 1982*.

We consider electrically conductive fluid in an external magnetic field. We observe the effect of electric fields and currents generation due to the fluid's motion. On the one hand the forces, which are able to significantly affect the motion, act upon the currents. On the other hand the currents do change the magnetic field. In this way the instability growth on the background of hydromagnetic and magnetic forces interaction is investigated. The curvature of the wall may be described by the curvilinear coordinates introduction.  

### Two-fluid model
The point we begin from is the full set of Maxwell's equations coupled with the Boltzmann equation describing the kinetic model of the fluid (in CGS unit):
\begin{equation*}
\frac{\partial f_\alpha}{\partial t}+\bar{u}\cdot(\bigtriangledown f_\alpha)+\frac{q_\alpha}{m_\alpha}(\bar{E}+\frac{1}{c}\bar{u}\times\bar{B})\frac{\partial f_\alpha}{\partial\bar{u}}=(\frac{\partial f_\alpha}{\partial t})_{col},\\
\bigtriangledown\cdot\bar{E}=4\pi Q,\\
\bigtriangledown\cdot\bar{B}=0,\\
\bigtriangledown\times\bar{E}=-\frac{1}{c}\frac{\partial\bar{B}}{\partial t},\\
\bigtriangledown\times\bar{B}=\frac{4\pi}{c}\bar{J}+\frac{1}{c}\frac{\partial\bar{E}}{\partial t},
\end{equation*}
where $f_\alpha(\bar{r},\bar{u},t)$ is the distribution function for each species. We apply $\alpha=i,e$.

According to the Boltzmann equation there are two forces acting on the particles: Lorentz force $q_\alpha(\bar{E}+\frac{1}{c}\bar{u}\times\bar{B})$, where $\bar{E}$ and $\bar{B}$ are the averaged values of electric and magnetic fields and the force representing collisions - it's introduced by the right hand of the equation. 

The appropriate moments for the Boltzmann equation are the mass $m_\alpha$, the momentum $m_\alpha\bar{u}$ and the kinetic energy $\frac{1}{2}m_\alpha u^2$. Putting them into the Boltzmann equation we get the equations for the conservation of mass, momentum and energy supplementing Maxwell's equations to the full system of the *two-fluid equations*:
\begin{equation*}
\frac{\partial n_\alpha}{\partial t}+\bigtriangledown\cdot{n_\alpha\bar{v}_\alpha}=\frac{d n_\alpha}{d t}+n_\alpha\bigtriangledown\cdot\bar{v}_\alpha=0,\\
n_\alpha m_\alpha\frac{d\bar{v}_\alpha}{d t}-q_\alpha n_\alpha(\bar{E}+\frac{1}{c}\bar{v}_\alpha\times\bar{B})+\bigtriangledown\bar{P}_\alpha=\bar{R}_\alpha,\\
\frac{3}{2}n\frac{d T_\alpha}{d t}+\bar{P}_\alpha\bigtriangledown\bar{v}_\alpha+\bigtriangledown\cdot\bar{h}=Q_\alpha,\\
\bigtriangledown\cdot\bar{E}=4\pi e(n_i-n_e),\\
\bigtriangledown\cdot\bar{B}=0,\\
\bigtriangledown\times\bar{E}=-\frac{1}{c}\frac{\partial\bar{B}}{\partial t},\\
\bigtriangledown\times\bar{B}=\frac{4\pi}{c}e(n_i\bar{v}_i-n_e\bar{v}_e)+\frac{1}{c}\frac{\partial\bar{E}}{\partial t},
\end{equation*}
where $n_\alpha$ is the macroscopic number density, $\bar{v}_\alpha$ is the fluid velocity, $p_\alpha$ is the scalar pressure, $\bar{P}_\alpha$ is the total pressure tensor, $\bar{\Pi}_\alpha$ is the anisotropic part of the pressure tensor, $\bar{h}_\alpha$ is the heat flux, $\bar{R}_\alpha$ represents the mean momentum transfer between the unlike particles due to the friction forces, $Q_\alpha$ is the heat generated due to the collisions.

### Single-fluid model
Now we make the approximations applicable for the physical problem we are going to solve. First is that we consider the non-relativistic case, so the term describing displacement currents $\frac{1}{c}\frac{\partial\bar{E}}{\partial t}$ can be neglected. Also, $\varepsilon_0=\frac{1}{4\pi c}\rightarrow 0$, so due to considering Maxwell's equations in SI we get that $\varepsilon_0\bigtriangledown\cdot\bar{E}$ can be neglected as well, what implies that $n_i=n_e$. Second is that we apply $m_e\rightarrow 0$ because of its smallnes compare to $m_i$. So, if we define so-called mass velocity, we get:
\begin{equation*}
\bar{U}=\frac{m_i n_i\bar{v}_i+m_e n_e\bar{v}_e}{m_i n_i+m_e n_e}\rightarrow\bar{U}=\bar{v}_i.
\end{equation*}
Apparently, the electron component velocity can be found from the equation:
\begin{equation*}
\bar{v}_e=\bar{U}-\frac{\bar{J}}{e n}.
\end{equation*}
We define the mass density as:
\begin{equation*}
\rho=n_i m_i,
\end{equation*}
consider that $\bar{R}_e+\bar{R}_i=0$ and finally the combination of the two first equaions leads us to the single-fluid equation:
\begin{equation*}
\rho\frac{d\bar{U}}{d t}-\frac{1}{c}\bar{J}\times\bar{B}+\bigtriangledown p=-\bigtriangledown\cdot\bar{\Pi}.
\end{equation*}
Due to the definition of the tensor $\bar{\Pi}$ we can write:
\begin{equation*}
\frac{\partial\Pi_{ik}}{\partial x_k}=\mu\frac{\partial^2 U_i}{\partial x_k^2}+\mu\frac{\partial}{\partial x_i}\frac{\partial U_k}{\partial x_k}-\mu\frac{2}{3}\frac{\partial}{\partial x_i}\frac{\partial U_l}{\partial x_l}+\epsilon\frac{\partial}{\partial x_i}\frac{\partial U_l}{\partial x_l}=\mu\frac{\partial^2}{\partial x_k^2}U_i+(\frac{\mu}{3}+\epsilon)\frac{\partial}{\partial x_i}\frac{\partial U_l}{\partial x_l},
\end{equation*}
where $\mu$ is the dynamic viscosity and $\epsilon$ is so-called "second viscosity".

Now the single-fluid equation takes form of the widely-known Navier-Stokes equation:
\begin{equation*}
\frac{d\bar{U}}{d t}=-\frac{1}{\rho}\bigtriangledown p+\nu\bigtriangleup\bar{U}+\frac{1}{\rho}(\frac{\mu}{3}+\epsilon)\bigtriangledown div\bar{U}-\frac{1}{4\pi\rho}\bar{B}\times curl\bar{B},
\end{equation*}
where $\nu=\frac{\mu}{\rho}$.
We consider the fluid to be incompressible:
\begin{equation*}
\rho=const\rightarrow div\bar{U}=0,\\
\frac{d\bar{U}}{d t}=-\frac{1}{\rho}\bigtriangledown p+\nu\bigtriangleup\bar{U}-\frac{1}{4\pi\rho}\bar{B}\times curl\bar{B}.
\end{equation*}
Due to the simple transformation we find that:
\begin{equation*}
\bar{B}\times curl\bar{B}=\bigtriangledown\frac{B^2}{2}-(\bar{B}\cdot\bigtriangledown)\bar{B},
\end{equation*}
thus, we obtain **the first MHD equation**:
\begin{equation*}
\frac{\partial\bar{U}}{\partial t}+(\bar{U}\cdot\bigtriangledown)\bar{U}=-\frac{1}{\rho}\bigtriangledown(p+\frac{B^2}{8\pi})+\frac{1}{4\pi\rho}(\bar{B}\cdot\bigtriangledown)\bar{B}+\nu\bigtriangleup\bar{U}.
\end{equation*}
Now we handle with the momentum two-fluid equation. In accordance of all of our presumtions and due to the finitude of the fluid's electrical conductivity the current density is introduced as:
\begin{equation*}
\bar{J}=\sigma(\bar{E}+\frac{1}{c}\bar{U}\times\bar{B}),
\end{equation*}
then due to the Ampere's law we can write:
\begin{equation*}
curl\bar{B}=\frac{4\pi}{c}{J}=\frac{4\pi}{c}\sigma(\bar{E}+\frac{1}{c}\bar{U}\times\bar{B}),\\
\bigtriangledown\times\bigtriangledown\times\bar{B}=\frac{4\pi}{c}\sigma(\bigtriangledown\times\bar{E}+\frac{1}{c}\bigtriangledown\times\bar{U}\times\bar{B}).
\end{equation*}
We use that $div\bar{B}=0$ and get:
\begin{equation*}
\bigtriangledown\times\bigtriangledown\times\bar{B}=-\bigtriangleup\bar{B}=\frac{4\pi\sigma}{c}(-\frac{1}{c}\frac{\partial\bar{B}}{\partial t}+\frac{1}{c}\bigtriangledown\times\bar{U}\times\bar{B}).
\end{equation*}
Thus, **the second MHD equation** takes form:
\begin{equation*}
\frac{\partial\bar{B}}{\partial t}=\bigtriangledown\times\bar{U}\times\bar{B}+\eta\bigtriangleup\bar{B},
\end{equation*}
where $\eta=\frac{c^2}{4\pi\sigma}$ is the magnetic diffusitivity (applied to be constant).

We have obtained an important simplification due to the neglecting the displacement currents - eliminating terms of the electric field $\bar{E}$. We are solving the system of equations in terms of $\bar{H}$ and the electric field $\bar{E}$ can be unequivocally defined from the solution.

## Initial system of equations
In *Liu's paper* they consider hydrodynamic case and solve the incompressible Navier-Stokes equations without external forces. We are going to solve MHD problem, so the magnetic forces must be reckoned in. The Navier-Stokes equation takes the next form:
\begin{equation*}
\frac{\partial\bar{U}}{\partial t}+(\bar{U}\cdot\bigtriangledown)\bar{U}=-\frac{1}{\rho}\bigtriangledown(p+\frac{B^2}{8\pi})+\frac{1}{4\pi\rho}(\bar{B}\cdot\bigtriangledown)\bar{B}+\nu\bigtriangleup\bar{U},
\end{equation*}
where $\bar{U}$ is the velocity, $\bar{B}$ is the magnetic field (we consider that $\bar{B}=\bar{H}$), $p$ is the pressure, $\rho$ is the density and $\nu$ is the kinematic viscosity.
To describe the behaviour of the magnetic field we use the next form of the induction equation:
\begin{equation*}
\frac{\partial\bar{B}}{\partial t}+(\bar{U}\cdot\bigtriangledown)\bar{B}=(\bar{B}\cdot\bigtriangledown)\bar{U}+\eta\bigtriangleup\bar{B},
\end{equation*}
where we consider $\bigtriangledown\cdot\bar{B}=0$ and $\eta$ (magnetic diffusivity) is constant.
Thuswise we get the system of equations in CGS units:
\begin{equation*}
\frac{\partial\bar{U}}{\partial t}+(\bar{U}\cdot\bigtriangledown)\bar{U}=-\frac{1}{\rho}\bigtriangledown\cdot(p+\frac{B^2}{8\pi})+\frac{1}{4\pi\rho}(\bar{B}\cdot\bigtriangledown)\bar{B}+\nu\bigtriangleup\bar{U},\\
\frac{\partial\bar{B}}{\partial t}+(\bar{U}\cdot\bigtriangledown)\bar{B}=(\bar{B}\cdot\bigtriangledown)\bar{U}+\eta\bigtriangleup\bar{B},\\
\bigtriangledown\cdot\bar{U}=0,\\
\bigtriangledown\cdot\bar{B}=0.
\end{equation*}
For convenience's sake we rewrite our system in HLU units:
\begin{equation*}
\frac{\partial\bar{U}}{\partial t}+(\bar{U}\cdot\bigtriangledown)\bar{U}=-\frac{1}{\rho}\bigtriangledown\cdot(p+\frac{B^2}{2})+\frac{1}{\rho}(\bar{B}\cdot\bigtriangledown)\bar{B}+\nu\bigtriangleup\bar{U},\\
\frac{\partial\bar{B}}{\partial t}+(\bar{U}\cdot\bigtriangledown)\bar{B}=(\bar{B}\cdot\bigtriangledown)\bar{U}+\eta\bigtriangleup\bar{B},\\
\bigtriangledown\cdot\bar{U}=0,\\
\bigtriangledown\cdot\bar{B}=0.
\end{equation*}

## Formulation of the eugenvalue problem
We find the components in cylindrical coordinates and get:
\begin{equation*}
\frac{\partial U_r}{\partial t}+(\bar{U}\cdot\bigtriangledown)U_r-\frac{U_\phi^2}{r}=\frac{1}{\rho}(\bar{B}\cdot\bigtriangledown)B_r-\frac{B_\phi^2}{\rho r}-\frac{\partial\Pi}{\partial r}+\nu(\bigtriangleup U_r-\frac{2}{r^2}\frac{\partial U_\phi}{\partial\phi}-\frac{U_r}{r^2}),\\
\frac{\partial U_\phi}{\partial t}+(\bar{U}\cdot\bigtriangledown)U_\phi+\frac{U_\phi U_r}{r}=\frac{1}{\rho}(\bar{B}\cdot\bigtriangledown)B_\phi+\frac{B_\phi B_r}{\rho r}-\frac{1}{r}\frac{\partial\Pi}{\partial\phi}+\nu(\bigtriangleup U_\phi+\frac{2}{r^2}\frac{\partial U_r}{\partial\phi}-\frac{U_\phi}{r^2}),\\
\frac{\partial U_z}{\partial t}+(\bar{U}\cdot\bigtriangledown)U_z=\frac{1}{\rho}(\bar{B}\cdot\bigtriangledown)B_z-\frac{\partial\Pi}{\partial z}+\nu\bigtriangleup U_z,\\
\frac{\partial B_r}{\partial t}+(\bar{U}\cdot\bigtriangledown)B_r=(\bar{B}\cdot\bigtriangledown)U_r+\eta(\bigtriangleup B_r-\frac{2}{r^2}\frac{\partial B_\phi}{\partial\phi}-\frac{B_r}{r^2}),\\
\frac{\partial B_\phi}{\partial t}+(\bar{U}\cdot\bigtriangledown)B_\phi=(\bar{B}\cdot\bigtriangledown)U_\phi-\frac{1}{r}(U_\phi B_r-U_r B_\phi)+\eta(\bigtriangleup B_\phi+\frac{2}{r^2}\frac{\partial B_r}{\partial\phi}-\frac{B_\phi}{r^2}),\\
\frac{\partial B_z}{\partial t}+(\bar{U}\cdot\bigtriangledown)B_z=(\bar{B}\cdot\bigtriangledown)U_z+\eta\bigtriangleup B_z,\\
\frac{\partial U_r}{\partial r}+\frac{U_r}{r}+\frac{1}{r}\frac{\partial U_\phi}{\partial\phi}+\frac{\partial U_z}{\partial z}=0,\\
\frac{\partial B_r}{\partial r}+\frac{B_r}{r}+\frac{1}{r}\frac{\partial B_\phi}{\partial\phi}+\frac{\partial B_z}{\partial z}=0,
\end{equation*}
where $\bar{U}=(U_r,U_\phi,U_z)$, $\bar{B}=(B_r,B_\phi,B_z)$ and $\Pi=\frac{1}{\rho}(p+\frac{B^2}{2})$.

Our next step is to linearize the system:
\begin{equation*}
U_r=U_r,\\
U_\phi=U_b+U_\phi,\\
U_z=U_z,\\
\Pi=\Pi^{'},\\
B_r=B_r,\\
B_\phi=B_\phi,\\
B_z=B_0+B_z,
\end{equation*}
where $U_b$ is the Blasius profile which needs to be calculated, $B_0$ is the constant magnetic field and $U_r, U_\phi, U_z, B_r, B_\phi, B_z$ are the velocity and the magnetic field components of pertrubation which are considered to be small quantities. We neglect the second order of smallness terms and apply the problem to be axisymmetric ($\frac{\partial}{\partial\phi}=0$) and find:
\begin{equation*}
\frac{\partial U_r}{\partial t}-\frac{2U_bU_\phi}{r}-\frac{U_b^2}{r}=\frac{B_0}{\rho}\frac{\partial B_r}{\partial z}-\frac{\partial\Pi^{'}}{\partial r}+\nu(\bigtriangleup U_r-\frac{U_r}{r^2}),\\
\frac{\partial U_\phi}{\partial t}+U_r(\frac{\partial U_b}{\partial r}+\frac{U_b}{r})=\frac{B_0}{\rho}\frac{\partial B_\phi}{\partial z}+\nu(\bigtriangleup U_\phi-\frac{U_\phi}{r^2}),\\
\frac{\partial U_z}{\partial t}=\frac{B_0}{\rho}\frac{\partial B_z}{\partial z}-\frac{\partial\Pi^{'}}{\partial z}+\nu\bigtriangleup U_z,\\
\frac{\partial B_r}{\partial t}=B_0\frac{\partial U_r}{\partial z}+\eta(\bigtriangleup B_r-\frac{B_r}{r^2}),\\
\frac{\partial B_\phi}{\partial t}=B_0\frac{\partial U_\phi}{\partial z}+B_r(\frac{\partial U_b}{\partial r}-\frac{U_b}{r})+\eta(\bigtriangleup B_\phi-\frac{B_\phi}{r^2}),\\
\frac{\partial B_z}{\partial t}=B_0\frac{\partial U_z}{\partial z}+\eta\bigtriangleup B_z,\\
\frac{\partial U_r}{\partial r}+\frac{U_r}{r}+\frac{\partial U_z}{\partial z}=0,\\
\frac{\partial B_r}{\partial r}+\frac{B_r}{r}+\frac{\partial B_z}{\partial z}=0.
\end{equation*}
We introduce the pertrubation components in the form:
\begin{equation*}
U_r=e^{\omega t}u(r)\cos(\alpha z),\\
U_\phi=e^{\omega t}v(r)\cos(\alpha z),\\
U_z=e^{\omega t}w(r)\sin(\alpha z),\\
B_r=e^{\omega t}\phi(r)\sin(\alpha z),\\
B_\phi=e^{\omega t}\psi(r)\sin(\alpha z),\\
B_z=e^{\omega t}\chi(r)\cos(\alpha z),\\
\Pi^{'}=e^{\omega t}\bar{w}(r)\cos(\alpha z),\\
\end{equation*}
and obtain:
\begin{equation*}
\nu(DD_*-\alpha^2-\frac{\omega}{\nu})u+\frac{\alpha B_0}{\rho}\phi+\frac{2U_b}{r}v-\frac{U_b^2}{r}=D\bar{w},\\
\nu(DD_*-\alpha^2-\frac{\omega}{\nu})v+\frac{\alpha B_0}{\rho}\psi=(D_* U_b)u,\\
\nu(DD_*-\alpha^2-\frac{\omega}{\nu})w-\frac{\alpha B_0}{\rho}\chi=-\alpha\bar{w},\\
\eta(DD_*-\alpha^2-\frac{\omega}{\eta})\phi=B_0\alpha u,\\
\eta(DD_*-\alpha^2-\frac{\omega}{\eta})\psi=B_0\alpha v-(\frac{d U_b}{d r}-\frac{U_b}{r})\phi,\\
\eta(DD_*-\alpha^2-\frac{\omega}{\eta})\chi=-B_0\alpha w,\\
D_* u=-\alpha w,\\
D_*\phi=\alpha\chi,
\end{equation*}
where $D$ and $D_*$ are the differential operators represented as (partial derivatives have no sense anymore, because we have only r-dependent unknown functions left):
\begin{equation*}
D=\frac{d}{d r},\\
D_*=\frac{1}{r}\frac{d}{d r}r.
\end{equation*}
Now we apply **the narrow gap approximation** (neglect all terms proportional to $\frac{1}{r}$ except for the one which is the centrifugal force), put the third equation into the first one. Also it's easy to show that with the narrow gap approximation the sixth equation equals to the fifth one, so we don't consider it and find:
\begin{equation*}
(D^2-\alpha^2-\frac{\omega}{\nu})(D^2-\alpha^2)u+\frac{\alpha B_0}{\nu\rho}(D^2-\alpha^2)\phi=\frac{2U_b\alpha^2}{\nu r}v,\\
\nu(D^2-\alpha^2-\frac{\omega}{\nu})v+\frac{\alpha B_0}{\rho}\psi=(D U_b)u,\\
\eta(D^2-\alpha^2-\frac{\omega}{\eta})\phi=B_0\alpha u,\\
\eta(D^2-\alpha^2-\frac{\omega}{\eta})\psi=B_0\alpha v-(D U_b)\phi,\\
D u=-\alpha w,\\
D\phi=\alpha\chi.
\end{equation*}
Finally we get the following system of equations:
\begin{equation*}
d^4 u-\frac{2U_b\alpha^2}{\nu r}v+\frac{\alpha B_0}{\nu\rho}d^2\phi=\frac{\omega}{\nu}d^2 u,\\
d^2-\frac{u}{\nu}D U_b+\frac{\alpha B_0}{\nu\rho}\psi=\frac{\omega}{\nu}v,\\
d^2-\frac{\alpha B_0}{\eta}u=\frac{\omega}{\nu}(\frac{\nu}{\eta})\phi,\\
d^2-\frac{\alpha B_0}{\eta}v+\frac{(D U_b)}{\eta}\phi=\frac{\omega}{\nu}(\frac{\nu}{\eta})\psi,\\
D u=-\alpha w,\\
D\phi=\alpha\chi,
\end{equation*}
where $d^2=(D^2-\alpha^2)$ and, as we remember, $D=\frac{d}{d r}$.
To solve our physical problem we switch from cylindrical coordinates to curvilinear in the following way:
\begin{equation*}
u\rightarrow -w_1,\\
v\rightarrow u_1,\\
w\rightarrow v_1,\\
\frac{d}{d r}\rightarrow -\frac{d}{d z},\\
r\rightarrow R-z,\\
\phi\rightarrow -\chi_1,\\
\psi\rightarrow\phi_1,\\
\chi\rightarrow\psi_1,
\end{equation*}
and obtain the system of equations:
\begin{equation*}
(\frac{d^2}{d z^2}-\alpha^2)^2w_1+\frac{2U_b\alpha^2}{\nu (R-z)}u_1+\frac{\alpha B_0}{\nu\rho}(\frac{d^2}{d z^2}-\alpha^2)\chi_1=\frac{\omega}{\nu}(\frac{d^2}{d z^2}-\alpha^2)w_1,\\
(\frac{d^2}{d z^2}-\alpha^2)u_1-\frac{1}{\nu}\left(\frac{d U_b}{d z}\right)w_1+\frac{\alpha B_0}{\nu\rho}\phi_1=\frac{\omega}{\nu}u_1,\\
(\frac{d^2}{d z^2}-\alpha^2)\chi_1-\frac{\alpha B_0}{\eta}w_1=\frac{\omega}{\nu}(\frac{\nu}{\eta})\chi_1,\\
(\frac{d^2}{d z^2}-\alpha^2)\phi_1-\frac{\alpha B_0}{\eta}u_1+\frac{1}{\eta}\left(\frac{d U_b}{d z}\right)\chi_1=\frac{\omega}{\nu}(\frac{\nu}{\eta})\phi_1,\\
\frac{d w_1}{d z}=-\alpha v_1,\\
\frac{d\chi_1}{d z}=\alpha\psi_1.
\end{equation*}
*R. Kobayashi* in his paper "*Instabilität laminarer Grenzschichten an konkaven Wänden im transversalen Magnetfeld,1969*" shows that $\chi_1\sim\frac{R_m}{Re}w_1$, so if we consider $R_m\ll Re$ we can neglect this term because of the second order of smallness. Besides, in this case:
\begin{equation*}
\frac{\omega}{\nu}(\frac{\nu}{\eta})\chi_1=\frac{\omega}{\nu}(\frac{R_m}{Re})\chi_1,\\
\frac{\omega}{\nu}(\frac{\nu}{\eta})\phi_1=\frac{\omega}{\nu}(\frac{R_m}{Re})\phi_1,
\end{equation*}
so we can neglect these terms too.
The system takes the following form:
\begin{equation*}
(\frac{d^2}{d z^2}-\alpha^2)^2w_1+\frac{2U_b\alpha^2}{\nu (R-z)}u_1+\frac{\alpha B_0}{\nu\rho}(\frac{d^2}{d z^2}-\alpha^2)\chi_1=\frac{\omega}{\nu}(\frac{d^2}{d z^2}-\alpha^2)w_1,\\
(\frac{d^2}{d z^2}-\alpha^2)u_1-\frac{1}{\nu}\left(\frac{d U_b}{d z}\right)w_1+\frac{\alpha B_0}{\nu\rho}\phi_1=\frac{\omega}{\nu}u_1,\\
(\frac{d^2}{d z^2}-\alpha^2)\chi_1=\frac{B_0\alpha}{\eta}w_1,\\
(\frac{d^2}{d z^2}-\alpha^2)\phi_1=\frac{B_0\alpha}{\eta}u_1,\\
\frac{d w_1}{d z}=-\alpha v_1,\\
\frac{d\chi_1}{d z}=\alpha\psi_1.
\end{equation*}
Then we express $\chi_1$ and $\phi_1$ in the first two equations in terms of $w_1$ and $u_1$ amd find:
\begin{equation*}
((\frac{d^2}{d z^2}-\alpha^2)^2+\frac{\alpha^2 B_0^2}{\rho\eta\nu})w_1+\frac{2U_b\alpha^2}{\nu (R-z)}u_1=\frac{\omega}{\nu}(\frac{d^2}{d z^2}-\alpha^2)w_1,\\
((\frac{d^2}{d z^2}-\alpha^2)^2+\frac{\alpha^2 B_0^2}{\rho\eta\nu})u_1-\frac{1}{\nu}\frac{d U_b}{d z}(\frac{d^2}{d z^2}-\alpha^2)=\frac{\omega}{\nu}(\frac{d^2}{d z^2}-\alpha^2)u_1,\\
(\frac{d^2}{d z^2}-\alpha^2)\chi_1=\frac{B_0\alpha}{\eta}w_1,\\
(\frac{d^2}{d z^2}-\alpha^2)\phi_1=\frac{B_0\alpha}{\eta}u_1,\\
\frac{d w_1}{d z}=-\alpha v_1,\\
\frac{d\chi_1}{d z}=\alpha\psi_1.
\end{equation*}
Now we have generalized eigenvalue problem:
\begin{equation*}
((\frac{d^2}{d z^2}-\alpha^2)^2+\frac{\alpha^2 B_0^2}{\rho\eta\nu})u_1-\frac{1}{\nu}\frac{d U_b}{d z}(\frac{d^2}{d z^2}-\alpha^2)w_1=\frac{\omega}{\nu}(\frac{d^2}{d z^2}-\alpha^2)u_1,\\
((\frac{d^2}{d z^2}-\alpha^2)^2+\frac{\alpha^2 B_0^2}{\rho\eta\nu})w_1+\frac{2U_b\alpha^2}{\nu (R-z)}u_1=\frac{\omega}{\nu}(\frac{d^2}{d z^2}-\alpha^2)w_1,
\end{equation*}
or:
\begin{equation*}
A\cdot\bar{V}=\omega\cdot B\cdot\bar{V},
\end{equation*}

## Boundary conditions
The boundary conditions on the velocity components: $u_1=0$ (no-slip condition), $v_1=0$ (the free stream velocity outside the boundary layer), $w_1=0$ (impermeability of the wall).
Now we formulate the boundary conditions on the magnetic field components:
#### 1) Insulating wall
In HLU units:
\begin{equation*}
\bigtriangledown\times\bar{B}=\frac{1}{c}\frac{\partial\bar{E}}{\partial t}+\frac{1}{c}\bar{J}.
\end{equation*}
For our problem we take an approximation $U\ll c$, so we can neglect the first term in the right hand of the equation. Consider the r-component:
\begin{equation*}
(\bigtriangledown\times\bar{B})_r=\frac{1}{c}J_r.
\end{equation*}
The condition of the insulating wall means that we have no current cross the wall, so we can write:
\begin{equation*}
(\bigtriangledown\times\bar{B})_r=0,\\
\frac{1}{r}\frac{\partial B_z}{\partial\phi}-\frac{\partial B_\phi}{\partial z}=0,
\end{equation*}
since we apply $\frac{\partial}{\partial\phi}=0$:
\begin{equation*}
\frac{\partial B_\phi}{\partial z}=0,\\
\alpha e^{\omega t}\psi\cos(\alpha z)=0,\\
\psi=0\rightarrow\phi_1=0.
\end{equation*}
#### 2) Perfectly conducting wall
At first we have no magnetic field across the wall, so $B_r=0\rightarrow e{\omega t}\phi\sin(\alpha z)=0\rightarrow\phi=0\rightarrow$
\begin{equation*}
\chi_1=0.
\end{equation*}
At second we have $J_z=0\rightarrow\frac{1}{r}(B_\phi+r\frac{\partial B_\phi}{\partial r}-\frac{\partial B_r}{\partial\phi})=0\rightarrow\psi=0\rightarrow$
\begin{equation*}
\frac{\partial\phi_1}{\partial z}=0.
\end{equation*}

Now we can formulate **the boundary conditions for the eigenvalue problem**: $u_1=\frac{d^2u_1}{d z^2}=w_1=\frac{d w_1}{d z}=0$.

## Limit of hydrodynamic case
It's easy to show that if we take the limit of $B_0=0$, we get the system of equations equivalent to the one from *Liu's paper*:
\begin{equation*}
(\frac{d^2}{d z^2}-\alpha^2)u_1-\frac{1}{\nu}\frac{d U_b}{d z}w_1=\frac{\omega}{\nu}u_1,\\
(\frac{d^2}{d z^2}-\alpha^2)^2w_1+\frac{2U_b\alpha^2}{\nu (R-z)}u_1=\frac{\omega}{\nu}(\frac{d^2}{d z^2}-\alpha^2)w_1,\\
\frac{d w_1}{d z}=-\alpha v_1,
\end{equation*}

with the boundary conditions: $u_1=v_1=w_1=0$

## Numerical calculations
### Blasius profile
The **Blasius profile** needs to be calculated before solving the eigenvalue problem. It leads us to the Blasius equation which can be solved numerically. Thus, our task is to define the opportune numerical method.
All of the known numerical methods can be divided into those that are explicit and into those that are implicit. Primarily our choice must proceed from the answering the question if our ODE-system is stiff or not. The explicit methods are not applicable if we are dealing with stiff system because of the rapid increasing of the number of solutions when the step size is sufficiently small. 

There's no universal definition of stiffness, but there's no actual need to check it analytically. I begin from the guess that the system isn't stiff. I use an explicit Runge-Kutta method of order 4 which is built into *scipy.integrate.ode()* function. Fortunatelly, the integrator has such a property that it's automatically interrupted when the rapid increasing of the number of solutions happens.

Accordong to the *Liu's paper*, $0\le\bar{z}\le93.75$ in the non-dimensional units (I take upper limit to be 66.3 because I apply the boundary layer thikness to be $\delta=\sqrt{\frac{2\nu x}{U_0}}$ and solve Blasius equation without factor 2). I consider longer interval: $0\le\bar{z}\le141,4$ (or $0\le\bar{z}\le100$ when $\delta=\sqrt{\frac{2\nu x}{U_0}}$).

We see that the integrator's interrupting, but it's easy to shorten the interval and make the conclusion that we are solving the stiff problem where the explicit method can be used until the variable reaches certain value. 

It leads us to the choice of the numerical method to solve this system. There are a lot of implicit methods providing different orders of accuracy. One of the most popular among them is so-called **Backward Differentiation Formulae** (*Uri M. Ascher and Linda R. Petzold."Computer Methods for Ordinary Differential Equations and Differential-Algebraic Equations", December 2, 1997, p. 131*). That's an implicit method which is characterized by that the solution is evaluated only at the right border of the current step. Another advantage of this method is the fact that it's built into the *scipy.integrate.odeint()* function which provides automatic switch from non-stiff system solving method to stiff system solving method (*Jeffrey J Heys."Chemical and biomedical engineering calculations using Python",2017, p. 205*). Besides, the function *scipy.integrate.odeint()* has simpler interface than *scipy.integrate.ode()* function does.

Thuswise and so, there is a code evaluating Blasius profile in the way which is not limited by the interval of certain maximal length.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

etalim=66.3
N=500

def f(y,t):
    y1,y2,y3=y
    return [y2,y3,-y3*y1]

t=np.arange(0,etalim,etalim/N)
y0=[0,0,0.469]

#built-in method BDF allows to solve stiff systems of differential equations
[y1,y2,y3]=odeint(f,y0,t,full_output=False).T

%matplotlib inline

fig=plt.figure(figsize=(8,6))
ax=fig.add_subplot(111)
ax.plot(t,y2,'-k',linewidth=2,label='$\\frac{d}{d \\eta} f{\\left (\\eta \\right )}$')
ax.plot(t,y3,'-g',linewidth=2,label='$\\frac{d^{2}}{d \\eta^{2}}  f{\\left (\\eta \\right )}$')
plt.ylabel('$\\frac{d}{d \\eta} f{\\left (\\eta \\right )}$,$\\frac{d^{2}}{d \\eta^{2}}  f{\\left (\\eta \\right )}$',\
           fontsize=18)
plt.xlabel('$\\eta$',fontsize=18)
plt.legend(fontsize=18,loc='center right')
plt.title('Blasius profile',fontsize=24)
ax.grid(True)
```

### Normalization
To solve the eqigenvalue problem we adimensionalise the equations:
\begin{equation*}
((\frac{d^2}{d z^2}-\alpha^2)^2+\frac{\alpha^2 B_0^2}{\rho\eta\nu})u_1-\frac{1}{\nu}\frac{d U_b}{d z}(\frac{d^2}{d z^2}-\alpha^2)w_1=\frac{\omega}{\nu}(\frac{d^2}{d z^2}-\alpha^2)u_1,\\
((\frac{d^2}{d z^2}-\alpha^2)^2+\frac{\alpha^2 B_0^2}{\rho\eta\nu})w_1+\frac{2U_b\alpha^2}{\nu (R-z)}u_1=\frac{\omega}{\nu}(\frac{d^2}{d z^2}-\alpha^2)w_1,
\end{equation*}
with:
\begin{equation*}
\bar{z}=\frac{z}{\delta},\quad\bar{U}=\frac{U_b}{U_0},\quad\bar{u}=\frac{u_1}{U_0},\quad\bar{w}=\frac{w_1}{U_0}Re,\\
\bar{\alpha}=\alpha\delta,\quad\bar{\omega}=\omega\frac{\delta^2}{\nu},\quad Re=\frac{U_0\delta}{\nu},\quad R_m=\frac{U_0\delta}{\eta},\\
\delta=\sqrt{\frac{2\nu x_0}{U_0}},\quad Ha=B_0\delta\sqrt{\frac{1}{\nu\eta\rho}},\quad G=\frac{U_0\delta}{\nu}\sqrt{\frac{\delta}{R}},
\end{equation*}
where $\delta$ is the boundary layer thickness, $U_0$ is the free stream velocity, $Re$ is the Reynolds number, $R_m$ is the magnetic Reynolds number, $Ha$ is the Hartmann number, $G$ is the Göertler number.

The normalized system takes the form:
\begin{equation*}
((\frac{d^2}{d\bar{z}^2}-\bar{\alpha}^2)^2+Ha^2\bar{\alpha}^2)\bar{u}-\frac{d\bar{U}}{d\bar{z}}(\frac{d^2}{d\bar{z}^2}-\bar{\alpha}^2)\bar{w}=\bar{\omega}(\frac{d^2}{d\bar{z}^2}-\bar{\alpha}^2)\bar{u},\\
((\frac{d^2}{d\bar{z}^2}-\bar{\alpha}^2)^2+Ha^2\bar{\alpha}^2)\bar{w}+2\bar{U}\bar{\alpha}^2G^2\bar{u}=\bar{\omega}(\frac{d^2}{d\bar{z}^2}-\bar{\alpha}^2)\bar{w},
\end{equation*}
with the boundary conditions:
\begin{equation*}
\bar{u}=0,\quad\frac{d^2\bar{u}}{d\bar{z}^2}=0,\\
\bar{w}=0,\quad\frac{d\bar{w}}{d\bar{z}}=0.
\end{equation*}

### Discretization 
To solve the eigenvalue problem we have to discretize the differential operator $\frac{d^2}{d\bar{z}^2}$. I use the finite difference to approximate the derivative and choose the second order of accuracy.

```python
from math import *
import numpy as np
import scipy.linalg as la
import matplotlib.pyplot as plt

%matplotlib inline

U0=500 #dimensional free-stream velocity [cm/s]
V=0.15 #dimensional kinetic viscosity [cm**2/s]
R=320 #dimensional radius of curvature [cm]
G=6 #Goertler number
alph=0.276 #normalized wavenumber as in Liu's paper

theta=(G**2*V**2*R/U0**2)**(1/3)#;print('theta',theta,sep='=') #momentum thickness
delta=theta*sqrt(2)/0.664#; print('delta',delta,sep='=') #boundary layer thickness if factor 2 is considered
#x0=(G*V**(1/4)*sqrt(R)/(U0**(1/4)*(0.664)**(3/2)))**(4/3); print('x0',x0,sep='=') #distance from the edge 
#delta=sqrt(2*V*x0/U0)
alpha=alph*delta/theta#;print('alpha',alpha,sep='=') #normalized wavenumber
h=etalim/N #step

K12=U0*delta/V #coefficient of matrix A12
K21=K12*2*alpha**2*h**4

A11=np.eye(N-2,N-2);A11=A11*(-2-alpha**2*h**2)
for k in range(1,N-2):
    A11[k][k-1]=1
    A11[k-1][k]=1
    
A12=np.zeros((N-2,N-4))
for k in range(0,N-4):
    A12[k+1][k]=y3[k+2]

A21=np.zeros((N-4,N-2))
for k in range(0,N-4):
    A21[k][k+1]=y2[k+2]/(R/delta-t[k+2])
    
A22=np.dot(A11,A11);A22=np.delete(A22,[0,N-3],0);A22=np.delete(A22,[0,N-3],1)

A12=A12*(-K12)*h**2;A21=A21*K21

A1=np.hstack((A11,A12));A2=np.hstack((A21,A22));A=np.vstack((A1,A2))

B11=np.eye(N-2,N-2);B11=B11*h**2*delta**2/V
B12=np.zeros((N-2,N-4));B21=np.zeros((N-4,N-2))
B22=A11*delta**2*h**2/V;
B22=np.delete(B22,[0,N-3],0);B22=np.delete(B22,[0,N-3],1)

B1=np.hstack((B11,B12));B2=np.hstack((B21,B22));B=np.vstack((B1,B2))

beta,v=la.eig(A,B)
v=np.real(v)
#beta>0 => growth of pertrubation
#beta<0 => decrease of pertrubation
#beta=0 => zero amplification

u=np.zeros((N,));w=np.zeros((N,))
for k in range(1,N-1):
    u[k]=v[k-1][np.argmax(beta)] #for fixed value of growth rate
for k in range(2,N-2):
    w[k]=v[k+N-4][np.argmax(beta)]
w[1]=w[2]/4
w[N-2]=w[N-3]/4

print('Ha=0')
print('G',G,sep='=')
print('alpha (normalized as in the paper)',alph,sep='=')
print('omeganorm',np.max(beta)*delta/(sqrt(2)*U0),sep='=')

fig=plt.figure(figsize=(16,8))
ax1=fig.add_subplot(121)
ax1.plot(t,u,'-k',linewidth=2,label='$u{\\left (\\eta \\right )}$')
ax1.plot(t,w,'-g',linewidth=2,label='$w{\\left (\\eta \\right )}$')
ax1.set_xlabel('$\\eta$',fontsize=18)
ax1.set_ylabel('Normalized eigenvector',fontsize=18)
ax1.set_title('Eigenvector components',fontsize=24)
ax1.grid(True)

ax2=fig.add_subplot(122)
ax2.plot(t,u,'-k',linewidth=2,label='$u{\\left (\\eta \\right )}$')
ax2.plot(t,w,'-g',linewidth=2,label='$w{\\left (\\eta \\right )}$')
ax2.set_xlabel('$\\eta$',fontsize=18)
ax2.set_ylabel('Normalized eigenvector',fontsize=18)
ax2.set_title('Eigenvector components (changed scale)',fontsize=24)
ax2.axis([-0.5,8,-0.1,0.3])
ax2.grid(True)
```
